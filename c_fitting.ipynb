```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 11, 0.1)
m = 2
b = 3
y = m*x + b
mu = 0
sigma = 0.2
noise =  np.random.normal(mu,sigma,x.shape)
yn = x+ noise

Y = y
X = np.ones((x.shape[0], 2))
X[:,0] = x

# inv(X'X)*X'Y
B = np.matmul(np.linalg.inv(np.matmul(X.T, X)), np.matmul(X.T, Y))

print(B)

fig, ax  = plt.subplots(1, 2, sharex='all', sharey='all', figsize=(12,12))
ax[0].scatter(x,yn)

# ax[0].plot(ax[0].get_xlim(), ax[0].get_xlim()*B[0]+ B[1])
print(ax[0].get_xlim()*B[0])

```
